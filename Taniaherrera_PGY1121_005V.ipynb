```python
class Departamento:
    def __init__(self, piso, tipo, precio_venta):
        self.piso = piso
        self.tipo = tipo
        self.estado = "Disponible"
        self.comprador = ""
        self.precio_venta = precio_venta

    def __str__(self):
        if self.estado == "Disponible":
            return f"{self.tipo}{self.piso} - Disponible"
        elif self.estado == "Vendido":
            return f"{self.tipo}{self.piso} - Vendido"

class MiCasaApp:
    def __init__(self):
        self.departamentos = []
        self.compradores = {}

    def iniciar_app(self):
        self.crear_departamentos()
        while True:
            self.mostrar_menu()
            opcion = input("Ingrese una opción: ")
            if opcion == "1":
                self.comprar_departamento()
            elif opcion == "2":
                self.mostrar_departamentos_disponibles()
            elif opcion == "3":
                self.mostrar_listado_compradores()
            elif opcion == "4":
                self.mostrar_ganancias_totales()
            elif opcion == "5":
                print("Gracias por utilizar Casafeliz App. ¡Hasta luego!")
                print("Tania Herrera 11 julio 2023")
                break
            else:
                print("Opción inválida. Por favor, ingrese una opción válida.")

    def crear_departamentos(self):
        for piso in range(1, 11):
            tipos_departamento = ["A", "B", "C", "D"]
            for tipo in tipos_departamento:
                precio_venta = self.obtener_precio(tipo, "V")

                departamento = Departamento(piso, tipo, precio_venta)
                self.departamentos.append(departamento)

    def obtener_precio(self, tipo, operacion):
        precios = {
            "A": {"V": 3.800,},
            "B": {"V": 3.000,},
            "C": {"V": 2.800,},
            "D": {"V": 3.500,},
        }
        return precios[tipo][operacion]

    def mostrar_menu(self):
        print("========== CasaFeliz App ==========")
        print("1. Comprar departamento")
        print("2. Mostrar departamentos disponibles")
        print("3. Ver listado de compradores")
        print("4. Mostrar ganancias totales")
        print("5. Salir")

    def comprar_departamento(self):
        print("============ Comprar departamento ============")
        piso = int(input("Ingrese el número del piso: "))
        tipo = input("Ingrese el tipo de departamento (A, B, C, D): ")
        departamento_seleccionado = None
        for departamento in self.departamentos:
            if departamento.piso == piso and departamento.tipo == tipo:
                departamento_seleccionado = departamento
                break

        if departamento_seleccionado is None:
            print("El departamento no existe. Por favor, ingrese nuevamente.")
            return

        if departamento_seleccionado.estado != "Disponible":
            print("El departamento seleccionado no está disponible.")
            return

        run = input("Ingrese el RUN (sin guiones ni puntos): ")
        departamento_seleccionado.comprador = run
        departamento_seleccionado.estado = "Vendido"
        self.compradores[run] = departamento_seleccionado

        print("Operación realizada correctamente.")

    def mostrar_departamentos_disponibles(self):
        print("========== Departamentos disponibles ==========")
        for departamento in self.departamentos:
            print(departamento)

    def mostrar_listado_compradores(self):
        print("========== Listado de compradores ==========")
        if self.compradores:
            for run in sorted(self.compradores.keys()):
                departamento = self.compradores[run]
                print(f"RUN: {run} - Departamento: {departamento}")
        else:
            print("No hay compradores registrados.")

    def mostrar_ganancias_totales(self):
        print("============ Ganancias totales =============")
        total_ventas = 0

        for departamento in self.departamentos:
            if departamento.estado == "Vendido":
                total_ventas += departamento.precio_venta

        print(f"Total ventas: ${total_ventas}")

App = MiCasaApp()
App.iniciar_app()
```
